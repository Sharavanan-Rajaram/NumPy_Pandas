**Q.no(3).Write a code to `MERGE`(`outer`)  data1 & data2 as `data`.** 

```python
data_1= pd.DataFrame({'EMP_ID':[2101,2102,2103,2104],'SALARY_YEAR1':[1000,2000,3000,4000],'SALARY_YEAR2':[2000,4000,6000,8000]})
data_1
```

```python
data_2= pd.DataFrame({'EMP_ID':[2101,2102,2103,2106],'SALARY_YEAR1':[3000,12000,16000,6000],'SALARY_YEAR2':[4000,15000,30000,8000]})
data_2
```

```python
data_outer=pd.merge(data_1,data_2,on='EMP_ID',how='outer')
data_outer
```

**Q.no(4).Write a code to `JOIN`(`inner`) data1 & data2 as `data11`.**
